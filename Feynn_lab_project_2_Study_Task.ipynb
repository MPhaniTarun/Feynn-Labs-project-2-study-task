<h2>Encoding categorical values using Label Encoder</h2> 

```python
# Doing label encoding for all the columns except for the Like column 
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for i in l:
    if i!="Like" and i!="Age":
        df[i]=le.fit_transform(df[i])
df
```

```python
df["Like"].unique()
```

```python
# Cleaning data in the Like column and conveting the categorical values of Like column to numeric values
import warnings
warnings.filterwarnings('ignore')

for i in range(len(df['Like'])):
    df['Like'][i]=int(df['Like'][i][-2::])
```

```python
df["Like"].replace(5,5, regex=True,inplace=True)
df["Like"].replace(4,5, regex=True,inplace=True)
df["Like"].replace(3,4, regex=True,inplace=True)
df["Like"].replace(2,4, regex=True,inplace=True)
df["Like"].replace(1,3, regex=True,inplace=True)
df["Like"].replace(0,3, regex=True,inplace=True)
df["Like"].replace(-1,3, regex=True,inplace=True)
df["Like"].replace(-2,2, regex=True,inplace=True)
df["Like"].replace(-3,2, regex=True,inplace=True)
df["Like"].replace(-4,1, regex=True,inplace=True)
df["Like"].replace(-5,1, regex=True,inplace=True)
```

```python
df['Like']=df.Like.astype(int)
```

<h1>K-Means Clustering</h1>

```python
X=np.array(df1)
```

```python
X
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
print(y_kmeans)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X[:,11],X[:,12],c=y_kmeans)
centers=kmeans.cluster_centers_
plt.scatter(centers[:,11],centers[:,12],c='red')
```

<p>After performing K-means clustering with 3 cluster groups, the above plot shown is for Age and Like columns with 3 clusters and 3 cluster centers. And we can infer from the plot is that  age group around 18 to 35 has a cluster center around 3.75 and age group around 36 to 52 has a cluster center around 3.2 and age group around 52 to 70 has a cluster center around 3.</p>

```python
# Data Distribution on graph 
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(40,40))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='DarkRed')
        ax.set_yscale('log')
    fig.tight_layout()  
    plt.show()

draw_histograms(df1,df1.columns,8,4)
```

# Calculating target segments

```python
df['Clusters']=kmeans.labels_
df
```

```python
df['VisitFrequency'] = le.fit_transform(df['VisitFrequency'])
visit_freq = df.groupby('Clusters')['VisitFrequency'].mean()
visit_freq = visit_freq.to_frame().reset_index()
visit_freq
```

```python
df['Like'] = le.fit_transform(df['Like'])
like = df.groupby('Clusters')['Like'].mean()
like = like.to_frame().reset_index()
like
```

```python
segment = like.merge(visit_freq, on='Clusters', how='left')
plt.figure(figsize = (5,4))
sb.scatterplot(x = "Like", y = "VisitFrequency",data=segment,s=400, color="purple")
plt.title("Simple segmentplot for this Mcdonalds dataset") 
plt.grid(True)
plt.xlabel("visit") 
plt.ylabel("like") 
plt.show()
```
